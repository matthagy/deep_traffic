```python
import os
import os.path
from glob import iglob
import re

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
parameter_names = ['side', 'ahead', 'behind', 'iterations', 'layers', 'layer_size']
```

```python
acc = []
for path in iglob('data/grid_search/*.log'):
    n = os.path.basename(path)
    parameters = list(map(int, n.replace('.log', '').split('_')))
    with open(path, 'rt') as fp:
        log = fp.read()
    m = re.search('MPH ([\d.]+)', log)
    if not m:
        continue
    mph = float(m.group(1))
    acc.append(parameters + [mph])
data = pd.DataFrame(acc, columns=parameter_names + ['trial', 'speed'])
```

```python
data.sort_values('speed', ascending=False)
```

```python
stat_data = data.groupby(parameter_names)['speed'].describe().reset_index()
stat_data = stat_data.sort_values('mean', ascending=False)
stat_data
```
